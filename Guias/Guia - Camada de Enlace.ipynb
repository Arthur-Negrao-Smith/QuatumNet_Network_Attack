# Guia da Camada de Enlace
### IMPORTANDO AS DEPENDÊNCIAS
##### O documento tem o objetivo de demostrar o funcionamento da camada de enlace, além de suas funções e ferramentas.

```python
from quantumnet.components import Network, Host
from quantumnet.objects import Qubit, Logger
```

##### Setando a topologia da rede.

```python
rede = Network()
rede.set_ready_topology("Grade", 3,3)
rede.draw()
Logger.activate(Logger)
```

##### Definindo os hosts alice e bob nos nós 1 e 2 respectivamente, e definindo os qubits que serão enviados.

```python
alice = rede.get_host(1)
bob = rede.get_host(2)
qubit1 = Qubit(1)
qubit2 = Qubit(2)
```

##### Ao enviar uma solicitação(request) para a rede, se a fidelidade dos qubits for superior a 0.5, a rede aceita a conexão e envia a chave. Caso contrário, uma nova tentativa de entrelaçamento é realizada. Se a tentativa falhar novamente, a rede tenta purificar o entrelaçamento para recuperar a conexão.

```python
rede.linklayer.request(1,2)
```

##### Verificando a fidelidade média dos pares EPRs criados na camada de Enlace.

```python
rede.linklayer.avg_fidelity_on_linklayer()
```

##### A função get_total_useds_qubits, será usada para mostrar quantos qubits foram usados na camada de Enlace.

```python
rede.get_total_useds_qubits()
```

##### Nesta função, get_total_used_erps, será calculado a quantidade de pares EPRs usados na camada de Enlace.

```python
rede.get_total_useds_eprs()
```

##### Verificando todos os canais, como seus pares EPRs.

```python
rede.get_eprs()
```

##### A função get_timeslot apresenta quantos timeslot se passaram durante a execução da camada.

```python
rede.get_timeslot()
```

##### No avg_fidelity_on_linklayer, é realizada o cálculo da média da fidelidade dos ERPs criados na camada de Enlace.

```python
rede.linklayer.avg_fidelity_on_linklayer()
```
